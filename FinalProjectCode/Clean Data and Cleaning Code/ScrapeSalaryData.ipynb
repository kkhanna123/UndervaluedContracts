```python
# Imports

from IPython.core.display import display, HTML

import pickle
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup, Comment
import matplotlib.pyplot as plt
import sys
import os
import time
from unidecode import unidecode
import math
```

```python
def scrapeSalaryData(year):
    # Players with similar names have uncommon url structures
    curNumber = 1

    brunson_url = 'https://hoopshype.com/salaries/players/'+str(year)+"-"+str(int(year)+1)+"/"
    
    # Use requests to retrieve data from a given URL
    brunson_response = requests.get(brunson_url)

    # Parse the whole HTML page using BeautifulSoup
    soup = BeautifulSoup(brunson_response.text, 'html.parser')

    list_header = []
    header = soup.find_all("table")[0].find("tr")
    
    for items in header:
        try:
            list_header.append(items.get_text().strip())
        except:
            continue
    
    # Data list
    data=[]
    
    # Table markers
    HTML_data = soup.find_all("table")[0].find_all("tr")[1:]
    
    for element in HTML_data:
        sub_data = []
        for sub_element in element:
            try:
                sub_data.append(sub_element.get_text().strip())
            except:
                continue
        data.append(sub_data)
        
    dataFrame = pd.DataFrame(data = data, columns = list_header)
    
    
    return dataFrame
```

```python
salaries = {}
for year in range(1994,2030):
    salaries[year] = scrapeSalaryData(year)

# Save salary data in a local dictionary so I don't have to scrape from source repeatedly
with open('dictOfSalaryData.pickle', 'wb') as f:
    pickle.dump(salaries, f)
```
