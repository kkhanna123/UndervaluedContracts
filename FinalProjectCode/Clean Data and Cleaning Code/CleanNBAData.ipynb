```python
from IPython.core.display import display, HTML

import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup, Comment
import matplotlib.pyplot as plt
import sys
import os
import time
from unidecode import unidecode
import math
```

```python
player_df = pd.read_csv('PlayerData.csv')

# Add padding at the end to make indexing easier 
columnsToUse = list(player_df.columns)
columnsToUse.append("DELETE_LATER_BOOKEND")

# Delete all columns with Unnamed (basically left over indices from merged dataframes)
ind = 0
while ind < len(columnsToUse):
    if "Unnamed" in columnsToUse[ind]:
        columnsToUse = columnsToUse[:ind]+columnsToUse[ind+1:]
        continue
    ind+=1
    
# Delete duplicate columns
ind = 0
while ind < len(columnsToUse):
    found = False
    for num in range(3,30):
        if str(num) in columnsToUse[ind]:
            columnsToUse = columnsToUse[:ind]+columnsToUse[ind+1:]
            found = True
            continue
            
    if not found:
        ind+=1

columnsToUse.remove('DELETE_LATER_BOOKEND')
            
player_df = player_df[columnsToUse]

display(player_df.head())
        
# Store Data in csv file to be used for future tasks
player_df.to_csv('CleanNBAData.csv',index=False)
```
